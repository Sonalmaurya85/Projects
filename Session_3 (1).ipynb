Q. The employee of the month function will return both the name and number of hours worked for the top performer (judged by number of hours worked).

work_hours = [('Abby',100),('Billy',400),('Cassie',800)]

Desired Output --> ('Cassie', 800)

___

```python
def max(work_hours1):
    max=0
    for i,j in work_hours1:
        if j>max:
            max=j
    return max

work_hours=[('Abby',100),('Billy',400),('Cassie',800)]
max=max(work_hours)
print(max)
```

```python
def find_employee_of_month(work_hours):
    """
    Given a dictionary of employees and their work hours, returns the name of the employee
    with the most hours worked, along with the number of hours worked by that employee.
    """
    employee_of_month = ''
    max_hours = 0
    
    for employee, hours in work_hours.items():
        if hours > max_hours:
            max_hours = hours
            employee_of_month = employee
    
    return (employee_of_month, max_hours)
```

```python
find_employee_of_month({'fred':800,'Billy':400,'fred':800, 'Cassie':801})
```

```python
class Circle:
    pi = 3.14

# Circle gets instantiated with a radius (default is 1)
    def __init__(self, radius=1):
        self.radius = radius 
        self.area = radius * radius * Circle.pi

# Method for resetting Radius
    def setRadius(self, new_radius):
        self.radius = new_radius
        self.area = new_radius * new_radius * Circle.pi

# Method for getting Circumference
    def getCircumference(self):
        return self.radius * self.pi * 2

```
